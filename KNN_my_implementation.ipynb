# K - Nearest Neighbours:

Creiamo dal principio l'algoritmo del K - Nearest Neighbours.

```python
# importazioni solite
import numpy as np
import pylab as pl
import scipy 
from scipy.spatial.distance import cdist
```

```python
#metto il seed dei valori random
np.random.seed(144)

# creiamo 100 punti in 2 dimensioni distribuiti gaussianamente con media 0
#e varianza 1
data = np. random.randn(100, 2)

#definiamo la mia label come un colore
#supponiamo che vi siano soltanto due valori, ossia due colori
label= np.random.randint(0, 2, 100)

label
```

```python
data.shape
```

```python
ax = pl.figure(figsize=(10,10)).add_subplot(1,1,1)
#faccio in modo che gli assi abbiano stessa dimensione
ax.axis('equal')

ax.plot(data[:,0],data[:,1], '.')
ax.set_xlabel("x")
ax.set_ylabel("y")

#plottato il mio feature space banale, che ho distribuito gaussianamente
```

```python
ax = pl.figure(figsize=(10,10)).add_subplot(1,1,1)
#faccio in modo che gli assi abbiano stessa dimensione
ax.axis('equal')

#plotto sottolineando i label 0 e i label 1

ax.plot(data[:,0][label==0],data[:,1][label==0], 'o')
ax.plot(data[:,0][label==1],data[:,1][label==1], 's')
ax.set_xlabel("x")
ax.set_ylabel("y")

#plottato il mio feature space banale, che ho distribuito gaussianamente
```

```python
#introduco una test particle

testparticle = np.random.randn(2)
testparticle
```

```python
ax = pl.figure(figsize=(10,10)).add_subplot(1,1,1)
#faccio in modo che gli assi abbiano stessa dimensione
ax.axis('equal')

#plotto sottolineando i label 0 e i label 1

ax.plot(data[:,0][label==0],data[:,1][label==0], 'o')
ax.plot(data[:,0][label==1],data[:,1][label==1], 's')
ax.plot(testparticle[0],testparticle[1], 'kx', ms=30, lw=3)
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
#costruisco il K nearest neighbours con la funzione cdist
cdist?
```

```python
#array delle distanze tra la testparticle e i dati
d = cdist(np.atleast_2d(testparticle), data)
d
```

```python
np.argmin(d)
d.shape
```

```python
#troviamo il minimo
np.min(d), d[:, np.argmin(d)]
```

```python
k = 1
labeltest = label[np.argmin(d)]
```

```python
ax = pl.figure(figsize=(10,10)).add_subplot(1,1,1)
#faccio in modo che gli assi abbiano stessa dimensione
ax.axis('equal')

#plotto sottolineando i label 0 e i label 1

colors=['blue','orange']
ax.plot(data[:,0][label==0],data[:,1][label==0], 'o')
ax.plot(data[:,0][label==1],data[:,1][label==1], 's')
ax.plot(testparticle[0],testparticle[1], 'kx', ms=30, lw=3, color=colors[labeltest])
ax.set_xlabel("x")
ax.set_ylabel("y")
```

La variabile random è stata messa di colore blu, poiché il suo primo vicino più vicino è un elemento di colore blu!

```python
# proviamo l'algoritmo con k = 5 nearest neighbours
k=5

#scelgo i 5 oggetti più vicini aventi distanza minore
nns = data[np.argsort(d)][0][:k]
nns
```

```python
nnslabel= label[np.argsort(d)][0][:k]
nnslabel
```

```python
#modo compatto di scrivere una proposizione logica
#se ho più labels 0 chiamalo 0
#se ho più labels 1 chiamalo 1
labeltest2 = 0 if (nnslabel == 0).sum() > (nnslabel == 1).sum() else 1
```

```python
ax = pl.figure(figsize=(10,10)).add_subplot(1,1,1)
#faccio in modo che gli assi abbiano stessa dimensione
ax.axis('equal')

#plotto sottolineando i label 0 e i label 1

colors=['blue','orange']
ax.plot(data[:,0][label==0],data[:,1][label==0], 'o')
ax.plot(data[:,0][label==1],data[:,1][label==1], 's')
ax.plot(testparticle[0],testparticle[1], 'kx', ms=30, lw=3, color=colors[labeltest2])
ax.plot(nns[:,0], nns[:,1], 'kx')
ax.set_xlabel("x")
ax.set_ylabel("y")
```

Nel modello a k=5 nearest neighbours abbiamo classificato che dato il nostro punto random, scelti i primi 5 vicini ho 3 vicini in blu e 2 vicini in arancione, quindi ho che la croce si classifica come BLU!
